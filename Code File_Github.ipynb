# Q1. How are the Crimes and Crime Rates in Chicago trending over the years?

```
# Using spark.sql to summarize total crimes by each year. Saving the result obtained to a temp table called Yearwise_Crimes

Yearwise_Crimes = spark.sql("select Year as Yr,count(ID) as Crimes from Crimes GROUP BY Year order by Year Asc")
Yearwise_Crimes.createOrReplaceTempView('Yearwise_Crimes')
Yearwise_Crimes.show()
```

```
# Joining the Yearwise_Crime table with the Population table using a left join with Year as the common variable. Then in the newly created table called Yearwise_Crime_Rates, adding a new column called Crime_Rate, created by using the Crimes and Population columns 

spark.sql("select CAST(Yr AS INT) as Yr, Crimes,Population.`Total Population` as population from Population inner join Yearwise_Crimes on Population.Year = Yearwise_Crimes.Yr order by Yr ASC").createOrReplaceTempView('Yearwise_Crime_Rates')
Yearwise_Crime_Rates = spark.sql("select Yr, Crimes,population,round(Crimes/population*100000,2) as Crime_Rate from Yearwise_Crime_Rates order by Yr ASC")
Yearwise_Crime_Rates.show()
```

# Q2. What is the nature of crimes being committed?

```
# Summarizing the total crimes by crime types across the last 20 years and storing the result to a table called Top_Crimes

spark.sql("select `Primary Type` as Crime_Type, count(`Primary Type`) as Crimes from Crimes GROUP BY Crime_Type order by Crimes DESC").createOrReplaceTempView('Top_Crimes')
Top_Crimes = spark.sql("select Crime_Type, Crimes,round((Crimes / (SELECT SUM(Crimes) FROM Top_Crimes)) * 100,2) AS Proportion from Top_Crimes order by Crimes DESC")
Top_Crimes.show(40, False)
```

```
# Summarizing the count of each crime by year

spark.sql("select Year, `Primary Type`, count(`Primary Type`) as Total from Crimes GROUP BY Year, `Primary Type` order by Year ASC").createOrReplaceTempView('Crime_by_year')

# Rerranging the data into the form of a pivot table with crime type as rows, year as column and count as values

pvt1=spark.sql("select * from (select `Primary Type`,Year,Total from Crime_by_year) PIVOT(sum(Total) for Year in (2015,2016,2017,2018,2019))")
pvt1.createOrReplaceTempView('pvt1')
spark.sql("select * from pvt1")
pvt2=spark.sql("select `Primary Type`, COALESCE(`2015`,0) as `2015`, COALESCE(`2016`,0) as `2016`, COALESCE(`2017`,0) as `2017`, COALESCE(`2018`,0) as `2018`, COALESCE(`2019`,0) as `2019` from pvt1")
pvt2.createOrReplaceTempView('pvt2')
spark.sql("select * from pvt2")

# Calculating the change in number of crimes committed each year for every type of crime and taking the average of the change in last 5 years

spark.sql("select `Primary Type`,round(((`2016`/`2015`)-1)*100,2) as rate_inc_2016, round(((`2017`/`2016`)-1)*100,2) as rate_inc_2017, round(((`2018`/`2017`)-1)*100,2) as rate_inc_2018, round(((`2019`/`2018`)-1)*100,2) as rate_inc_2019 from pvt2 ").createOrReplaceTempView('pvt3')
spark.sql("select `Primary Type`, `rate_inc_2016`, `rate_inc_2017`, `rate_inc_2018`, `rate_inc_2019`, round((`rate_inc_2016`+`rate_inc_2017`+`rate_inc_2018`+`rate_inc_2019`)/4,2) as avg from pvt3").createOrReplaceTempView('pvt4')

# Top 5 Crimes Increasing the most over the last 5 years
spark.sql("select `Primary Type`, `rate_inc_2016`, `rate_inc_2017`, `rate_inc_2018`, `rate_inc_2019`, avg from pvt4 ORDER BY avg DESC limit 5").show(40, False)

# Top 5 Crimes which have gone down the most in the last 5 years
spark.sql("select `Primary Type`, `rate_inc_2016`, `rate_inc_2017`, `rate_inc_2018`, `rate_inc_2019`, avg from pvt4 ORDER BY avg ASC limit 5").show(40, False)
```

```
# Creating two tables, one each for total crimes by year for prostitution and total crimes by year for secual assault. Then joining the two tables by Year to create a common table and stroring results as Corr. Using the Yearwise crimes for prostitution and sexual assault, the correlation between the two crimes is calculated

spark.sql("select Year as Yr,count(*) as Prostitution from Crimes where `Primary Type` = 'PROSTITUTION' group by Yr").createOrReplaceTempView('Prostitution')
spark.sql("select Year,count(*) as Assault from Crimes where `Primary Type` = 'CRIMINAL SEXUAL ASSAULT' group by Year").createOrReplaceTempView('Sexual_Assault')
Corr = spark.sql("Select Yr,Prostitution,Sexual_Assault.Assault from Prostitution inner join Sexual_Assault on Prostitution.Yr=Sexual_Assault.Year order by Yr ASC")
print("Correlation between Prostitution and Sexual Assault is"+" "+str(Corr.stat.corr('Prostitution', 'Assault')))
```

# Q4. When do most crimes occur?

```
# Extracting months from the date field and summarizing total crimes by Months and storing the output in a table called Monthwise_Crimes

spark.sql("select CAST(LEFT(Date, 2) AS INT) as Month,count(*) as Crimes from Crimes GROUP BY Month").createOrReplaceTempView('Monthwise_Crimes')
Monthwise_Crimes = spark.sql("select Month,Crimes,round((Crimes / (SELECT SUM(Crimes) FROM Monthwise_Crimes)) * 100,2) AS Proportion from Monthwise_Crimes")
Monthwise_Crimes.show()
```

```
# Extracting time from the date field, categorizing the times into dawn, morning, afternoon, evening and night and then finally summarizing total crimes by Time of the Day and storing the output in a table called Timewise_Crimes

spark.sql("select substring(Date,12,23) as Time,count(*) as Crimes from Crimes group by Time").createOrReplaceTempView('Timewise_Crimes')
a = spark.sql("select CAST(LEFT(Time, 2) AS INT) as Hour, substring(Time,9,11) as Time,Crimes from Timewise_Crimes")
from pyspark.sql.functions import when,col
a = a.withColumn("Time_of_day",when((col("Hour")>=1) & (col("Hour")<=6) & (col("Time")==" AM"),"Dawn").when((col("Hour")>6) & (col("Hour")<12) & (col("Time")==' AM'),"Morning").when((col("Hour")>=1) & (col("Hour")<=4) & (col("Time")==' PM'),"Afternoon").when((col("Hour")>4) & (col("Hour")<=8) & (col("Time")==' PM'),"Evening").when((col("Hour")>8) & (col("Hour")<=11) & (col("Time")==' PM'),"Night").when((col("Hour")==12) & (col("Time")==' PM'),"Afternoon").when((col("Hour")==12) & (col("Time")==' AM'),"Night"))
a.createOrReplaceTempView('Timewise_Crimes')
spark.sql("select Time_of_Day,sum(Crimes) AS Crimes from Timewise_Crimes group by Time_of_Day").createOrReplaceTempView('Timewise_Crimes')
Timewise_Crimes = spark.sql("select Time_of_Day,Crimes,round((Crimes / (SELECT SUM(Crimes) FROM Timewise_Crimes)) * 100,2) AS Proportion from Timewise_Crimes")
Timewise_Crimes.show()
```

```
# Extracting dates from the date field, categorizing the dates into starting, mid and ending of the month and then finally summarizing total crimes by Time_of_Month and storing the output in a table called Datewise_Crimes

spark.sql("select substring(Date,4,6) as Date,count(*) as Crimes from Crimes group by Date").createOrReplaceTempView('Datewise_Crimes')
b = spark.sql("select CAST(LEFT(Date, 2) AS INT) as Date, Crimes from Datewise_Crimes")
b = b.withColumn("Time_of_Month",when((col("Date")>=1) & (col("Date")<=10),"Starting").when((col("Date")>10) & (col("Date")<=20),"Mid").when((col("Date")>20) & (col("Date")<=31),"End"))
b.createOrReplaceTempView('Datewise_Crimes')
spark.sql("select Time_of_Month,sum(Crimes) AS Crimes from Datewise_Crimes group by Time_of_Month").createOrReplaceTempView('Datewise_Crimes')
Datewise_Crimes = spark.sql("select Time_of_Month,Crimes,round((Crimes / (SELECT SUM(Crimes) FROM Datewise_Crimes)) * 100,2) AS Proportion from Datewise_Crimes")
Datewise_Crimes.show()
```

# Q5. How is the response by law enforcement authorities?

```
# Summarizing the number of arrests made by Year and saving to a temp table called Arrests

Arrests = spark.sql("select Year as Yr,count(Arrest) as Overall_Arrest from Crimes where Arrest= 'true' GROUP BY Year,Arrest order by Year ASC")
Arrests.createOrReplaceTempView('Arrests')
Arrests.show()
```

```
# Creating another table called Arrest_Percent, where we calculate the Overall Criminal Cases per Year. Then inner joining the Arrests and Arrest_Percent table on Year. From the result obtained we calculate the percentage of arrests made out of the total criminal cases each year 

spark.sql("select Year as years, count(ID) as cases_per_year from Crimes GROUP BY Year order by Year ASC").createOrReplaceTempView('Arrest_Percent')
spark.sql("select Yr ,Overall_Arrest,Arrest_Percent.cases_per_year from Arrests inner join Arrest_Percent on Arrest_Percent.years= Arrests.Yr order by yr ASC").createOrReplaceTempView('Arrest_Percent')
Arrest_Percent = spark.sql("select Yr,Overall_Arrest,cases_per_year, round((Overall_Arrest/cases_per_year)*100,2) as Percent from Arrest_Percent order by Yr ASC")
Arrest_Percent.show()
```

```
# Creating a new table called Arrest_by_Crime1, where we calculate the crimes by crime type. Another table called Arrest_by_Crime2 is created summarizing the arrests made by each crime type.

spark.sql("select distinct(`Primary Type`) as Crimes,count(`Primary Type`) as cases_per_crime from Crimes GROUP BY `Primary Type`").createOrReplaceTempView('Arrest_by_Crime1')
spark.sql("select `Primary Type`,count(Arrest) as Overall_Arrest_Crime from Crimes where Arrest= 'true' GROUP BY `Primary Type`,Arrest").createOrReplaceTempView('Arrest_by_Crime2')

# Inner joining the Arrests and Arrest_Percent table on Year

spark.sql("select `Primary Type`,Overall_Arrest_Crime,Arrest_by_Crime1.cases_per_crime from Arrest_by_Crime2 inner join Arrest_by_Crime1 on Arrest_by_Crime1.Crimes= Arrest_by_Crime2.`Primary Type` order by `Primary Type` ASC").createOrReplaceTempView('Arrest_by_Crime')

# From the result obtained we calculate the percentage of arrests made out of the total criminal cases each year 

# Identifying the top 5 crime types with highest arrest % 

spark.sql("select `Primary Type`,Overall_Arrest_Crime,cases_per_crime, round((Overall_Arrest_Crime/cases_per_crime)*100,2) as Percent from Arrest_by_Crime order by Percent DESC").show(5)

# Identifying the top 5 crime types with lowest arrest % 

spark.sql("select `Primary Type`,Overall_Arrest_Crime,cases_per_crime, round((Overall_Arrest_Crime/cases_per_crime)*100,2) as Percent from Arrest_by_Crime order by Percent ASC").show(5)
```

```
# Extracting dates from the Date and Updated On column and converting into the appropriate date format. Saving the results to a temp table called Investigation_Length

from pyspark.sql.functions import col, unix_timestamp, to_date,datediff

spark.sql("select LEFT(Date, 10) as Crime_Date,LEFT(`Updated On`, 10) as Updated_Date,Year,`Primary Type` as Crime_Type,District from Crimes").createOrReplaceTempView('Investigation_Length')
spark.sql("select to_date(cast(unix_timestamp(Crime_Date, 'MM/dd/yyyy') as TIMESTAMP)) as Crime_Date,to_date(cast(unix_timestamp(Updated_Date, 'MM/dd/yyyy') as TIMESTAMP)) as Updated_Date,Year, Crime_Type,District from Investigation_Length").createOrReplaceTempView('Investigation_Length')

# Calculating the difference between the updated date and the crime date to obtain the length of investigation in a new column called Length_of_Investigation

spark.sql("select Crime_Date,Updated_Date,Year,Crime_Type,District,DATEDIFF(Updated_Date, Crime_Date) AS Length_of_Investigation from Investigation_Length").createOrReplaceTempView('Investigation_Length')
LoI = spark.sql("select * from Investigation_Length")
LoI.show()
```

```
# Summarizing the average of the investigation length by each year

Inv_by_year = spark.sql("select Year,sum(Length_of_Investigation)/count(Length_of_Investigation) as Length_of_Investigation from Investigation_Length group by Year order by Year")
#Inv_by_year.show()
display(Inv_by_year)
```

```
# Summarizing the average of the investigation length by each crime type and showing the top and bottom  5 crime types in terms of average lenght of investigation over the years

spark.sql("select Crime_Type,round(sum(Length_of_Investigation)/count(Length_of_Investigation),0) as Length_of_Investigation from Investigation_Length group by Crime_Type order by Length_of_Investigation DESC limit 5").show(40,False)
spark.sql("select Crime_Type,round(sum(Length_of_Investigation)/count(Length_of_Investigation),0) as Length_of_Investigation from Investigation_Length group by Crime_Type order by Length_of_Investigation ASC limit 5").show(40,False)
```
